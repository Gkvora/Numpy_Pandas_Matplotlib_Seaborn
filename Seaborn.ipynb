# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Line plot

```python
a=[1,2,3,4,5,6,7,8,9]
b=[2,4,2,6,7,8,4,6,7]
```

```python
# from matplotlib
#plt.plot(a,b)
#plt.show() (optional)
```

```python
x1=pd.DataFrame({"a":a,"b":b})
x1
```

```python
sns.lineplot(x="a",y="b",data=x1)
# plt.show() ------ graph par ak line aave te remove karva
```

```python
sns.lineplot(x=a,y=b)
```

```python
y1=sns.load_dataset("Iris")
y1
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=y1)
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=y1,hue="species")
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=y1,hue="species",size=5)
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=y1,hue="species",style="species")
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=y1,hue="species",style="species",palette="Accent")
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=y1,hue="species",style="species",palette="Accent",legend="full")
```

### Histogram

```python
y1
```

```python
sns.displot(y1["sepal_length"])
```

```python
# Different parameter
sns.displot(y1["sepal_length"],bins=[4,5,6,7,8])
```

```python
sns.displot(y1["sepal_length"],bins=[4,5,6,7,8],kde=True)
```

```python
sns.displot(y1["sepal_length"],bins=[4,5,6,7,8],kde=True,rug=True)
```

```python
sns.displot(y1["sepal_length"],bins=[4,5,6,7,8],kde=True,rug=True,color="g")
```

```python
sns.displot(y1["sepal_length"],kde=True,rug=True,color="g",log_scale=True)
```
