# Codebasics (Numpy & Pandas)

```python
import numpy as np
```

```python
a=np.arange(12).reshape(3,4)
```

```python
a
```

```python
for i in a:
    print(i)
```

```python
for i in a:
    for j in i:
        print(j)
```

```python
for i in a.flat:
    print(i)
```

```python
for x in np.nditer(a,order='c'):
    print(x)
```

```python
for x in np.nditer(a,order='f'):# f=fortran
    print(x)
```

```python
for x in np.nditer(a,order='f',flags=['external_loop']):# f=fortran
    print(x)
```

```python
for x in np.nditer(a,op_flags=['readwrite']):
    x[...]=x*x
```
